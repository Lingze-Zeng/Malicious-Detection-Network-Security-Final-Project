```python
import requests
import pandas as pd
import numpy as np

r = requests.get(url='https://corporationwiki.com/Ohio/Columbus/frank-s-benson-P3333917.aspx')
print(r)
print(type(r))
```

```python
df = pd.read_csv('malicious_phish.csv')
MURL_NPlist = df.to_numpy()
# print(MURL_NPlist[:,1])
new_dim = np.zeros((MURL_NPlist.shape[0],2))
# print(new_dim)
# print(new_dim.shape[0])
pre_shaped_MURL_NPlist = np.append(MURL_NPlist,new_dim,axis=1)
i = 0
active_url = 0
bengin_url = 0
defacement_url = 0
phishing_url = 0
while i < 100:
    try:
        try:
            j=0
            while j<3:
                try:
                    r = requests.get(url=pre_shaped_MURL_NPlist[i,0],timeout=5)
                    break
                except requests.exceptions.Timeout:
                    j+=1
        except requests.exceptions.MissingSchema:
            j=0
            while j<3:
                try:
                    r = requests.get(url = 'http://' + pre_shaped_MURL_NPlist[i,0],timeout=5)
                    break
                except requests.exceptions.Timeout:
                    j+=1
        pre_shaped_MURL_NPlist[i,2] = r.status_code
        # print(r.status_code)
    except requests.RequestException as e:
        i+=1
        # print(e)
        continue
    if r.status_code == 200:
        pre_shaped_MURL_NPlist[i,3] = 1
        active_url+=1
        if pre_shaped_MURL_NPlist[i,1] == 'benign':
            bengin_url+=1
        elif pre_shaped_MURL_NPlist[i,1] == 'defacement':
            defacement_url+=1
        elif pre_shaped_MURL_NPlist[i,1] == 'phishing':
            phishing_url+=1
    i+=1
    print(i, end='\r')
print(pre_shaped_MURL_NPlist[0:i-1,:])
print('active:     ', active_url)
print('benign:     ', bengin_url)
print('defacement: ', defacement_url)
print('phishing:   ', phishing_url)

```
